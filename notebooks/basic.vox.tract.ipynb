# Voxels and streamlines

Voxel-based volumes and triangle-based meshes can exist together in the same scene. NiiVue considers tractography streamlines as a special class of meshes. Note that the `mesh_xray` effect allows you to visualize meshes that are beneath the surface of the volume rendering. The more advanced tracts notebook illustrates ways that you can modify streamlines appearance.

```
from ipyniivue import NiiVue, ShowRender

nv = NiiVue()
nv.load_volumes([{"path": "../images/mni152.nii.gz"}])
nv.load_meshes([
        {"path": "../images/dpsv.trx"},
     ])
nv.opts.multiplanar_show_render = ShowRender.ALWAYS
nv.set_clip_plane(-0.1, 270, 0)
nv.opts.mesh_xray = 0.02
nv
```
