# Minimal ipyniivue demo

This notebook demonstrates the minimal usage of `NiiVue` loading a single local NIfTI image. The [documentation](https://niivue.github.io/ipyniivue/) describes additional features.


```
from ipyniivue import NiiVue

nv = NiiVue()
nv.load_volumes([{'path': '../images/mni152.nii.gz'}])
nv
```
