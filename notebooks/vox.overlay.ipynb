# Voxel overlay

This notebook loads a background image (a T1-weighted MRI) as well as and overlay image of the hippocampus. Note that you can set some properties (such as colormap and opacity) both when you load an image, or subsequently. This also illustrates assigning custom keys to change the view.

```
from ipyniivue import NiiVue, SliceType

# based on https://niivue.github.io/niivue/features/basic.multiplanar.html

volumes = [
    {
        "path": "../images/mni152.nii.gz",
        "colormap": "gray",
        "visible": True,
        "opacity": 1.0,
    },
    {
        "path": "../images/hippo.nii.gz",
        "colormap": "red",
        "visible": True,
        "opacity": 1,
    },
]
nv = NiiVue(slice_type=SliceType.MULTIPLANAR)
nv.load_volumes(volumes)


nv.volumes[0].opacity = 0.3
nv.volumes[0].colorbar_visible = False
nv.volumes[1].colormap = "blue"

nv.opts.show_3d_crosshair = True
nv.opts.crosshair_color = [0, 1, 1, 1]
nv.opts.back_color = [1, 1, 1, 1]
nv.opts.clip_plane_color = [0, 1, 1, 1]
nv.opts.is_colorbar = True
nv.opts.view_mode_hot_key = "KeyN"

nv
```
